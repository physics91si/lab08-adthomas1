```python
import numpy as np
from scipy import integrate
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("droptower_vdata.txt",unpack = True)
time = data[0]
velocity = data[1]
f, (p,v,a) = plt.subplots(1,3)
plt.gcf().set_size_inches(15,5)
v.scatter(time,velocity,marker = 'o')
displacement = (integrate.cumtrapz(velocity,time,initial=0))
acceleration = np.diff(velocity)
p.scatter(time,displacement, marker = 'o')
a.scatter(time[0:len(time)-1],acceleration, marker = 'o')
p.set_title('Position')
v.set_title('Velocity')
a.set_title("Acceleration")
plt.show()

```

```python
apos = acceleration
total = 0
number = 0
for i in apos:
    if (i>=0):
        total= total + i
        number = number + 1
average = total / number
print (average)
```
